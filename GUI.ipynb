```python
#@title Xử lý đầu vào
#Sex
if Sex == "Male": Sex = 1
else: Sex = 0
# Age
Age = float(Age)
# Current Smoker
if Current_Smoker == "Yes": Current_Smoker = 1
else: Current_Smoker = 0
# Cigs Per Day
Cigs_Per_Day = float(Cigs_Per_Day)
# BP Meds
if BP_Meds == "Yes": BP_Meds = 1
else: BP_Meds = 0
# Prevalent Stroke
if Prevalent_Stroke == "Yes": Prevalent_Stroke = 1
else: Prevalent_Stroke = 0
# Prevalent Hyp
if Prevalent_Hyp == "Yes": Prevalent_Hyp = 1
else: Prevalent_Hyp = 0
# Diabetes
if Diabetes == "Yes": Diabetes = 1
else: Diabetes = 0
# Education
if Education == "Some high school": Education = 1
elif Education == "high school/GED": Education = 2
elif Education == "some college/vocational school": Education = 3
else: Education = 4
# Tot Chol
Tot_Chol = float(Tot_Chol)
# Sys BP
Sys_BP = float(Sys_BP)
# Dia BP
Dia_BP = float(Dia_BP)
# BMI
BMI = float(BMI)
# Heart Rate
Heart_Rate = float(Heart_Rate)
# Glucose
Glucose = float(Glucose)
```

```python
import pandas as pd
import numpy as np
def transformer(data, columns):
  df = pd.DataFrame(data, columns=columns)
  return df

df = transformer(np.array([[Sex, Age, Education, Current_Smoker, Cigs_Per_Day,
                            BP_Meds, Prevalent_Stroke, Prevalent_Hyp, Diabetes,
                            Tot_Chol, Sys_BP, Dia_BP, BMI, Heart_Rate, Glucose]]),
 ["male", "age", "education", "currentSmoker", "cigsPerDay", "BPMeds",
  "prevalentStroke", "prevalentHyp", "diabetes", "totChol", "sysBP", "diaBP",
  "BMI", "heartRate", "glucose"])
```

```python
df_mean = [49.498332, 9.468495, 236.647065, 130.725340,
           82.116997, 25.603182, 75.443329, 79.591493]
df_std = [ 8.4929893,  11.64949992, 40.720829, 20.66544055,
          11.6539426, 3.56207153, 11.11330976, 13.65492183]

#data from mean_std.txt saved Nhom1.ipynb
def scaler(data, float_columns, df_mean, df_std):
  data[float_columns] = (data[float_columns] - df_mean) / df_std
  return data

data = scaler(df, ['age', 'cigsPerDay', 'totChol', 'sysBP',
                   'diaBP', 'BMI', 'heartRate', 'glucose'], df_mean, df_std)
print(data)
```

```python
# Tải mô hình Logistic Regression từ file
import pickle
with open('logistic_model.pkl', 'rb') as file:
    logistic_model = pickle.load(file)

# Thực hiện dự đoán
prediction = logistic_model.predict(data)
probability = logistic_model.predict_proba(data)[0][1]

# Hiển thị kết quả
result = "Có bệnh" if prediction == 1 else "Không bệnh"
print(f"Dự đoán: {result}")
print(f"Xác suất mắc bệnh: {probability*100}%")
```
